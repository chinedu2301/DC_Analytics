```python
# Import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import seaborn as sns
%matplotlib inline
```

```python
# Load the data from github repo:

url = "https://raw.githubusercontent.com/chinedu2301/DC_Analytics/main/mass_shooting_2020_usa.csv"
mass_shooting_df = pd.read_csv(url, parse_dates = ['Incident Date']) # parse_dates converts the week_end column to datetime
mass_shooting_df.head(n = 10)
```

```python
# Check to see the data type of each variable
mass_shooting_df.info()
```

```python
# Check the summary statistics
mass_shooting_df.describe().T
```

```python
# replace all the spaces in the column names with underscores
mass_shooting_df.columns  = [c.replace(' ', '_') for c in mass_shooting_df.columns]
mass_shooting_df.columns
```

```python
# rename the columns
req_col_new_name = {'Incident_ID':'incident_id', 'Incident_Date':'incident_date', 'State':'state',
                    'City_Or_County': 'city', 'Address':'address', '#_Killed':'no_killed', '#_Injured':'no_injured'}
mass_shooting_df = mass_shooting_df.rename(columns = req_col_new_name)
mass_shooting_df.head(n = 10)
```

```python
# filter the required columns
req_col = ['incident_id', 'incident_date', 'state', 'city', 'address', 'no_killed', 'no_injured']
mass_shooting = mass_shooting_df[req_col]
mass_shooting.head(n = 10)
```

```python
# Write the cleaned dataset to csv
mass_shooting.to_csv('mass_shooting2020_usa_clean.csv', index = False)
```
