```python
# Import required libraries
import pandas as pd
import numpy as np
from requests import get
from bs4 import BeautifulSoup as bs
```

```python
# urls to scrape
uri = "https://www.kdnuggets.com/2017/05/42-essential-quotes-data-science-thought-leaders.html"
url = [uri, uri+'/2']
url
```

## Scrape the first page of the website

#### Access the website

```python
page = get(uri)
page
```

#### Use BeautifulSoup to parse the response from the webserver

```python
soup = bs(page.content, parser = "html5lib")
```

```python
name = soup.find(class_ = 'post-66316')
name.find_all('li')
```

```python
# Get the author of quotes from the first page
name = soup.find(class_ = 'post-66316')
links = name.find_all('a')
author = [item.text for item in links][2::2]
author
```

```python
# Get the quotes for the first page
name = soup.find(class_ = 'post-66316')
quote_list = name.find_all('li')
quote = [quote.text for quote in quote_list]
quote
```

#### Combine the author and quotes of the first page into a dataframe

```python
quotes_author = pd.DataFrame({
    'Quote': quote,
    'Author': author
})
quotes_author
```

## Scrape the second page of the website

```python
url2 = uri + "/2"
url2
```

```python
# make request to the webserver for the second page
rep2 = get(url2).content
soup2 = bs(rep2, 'html5lib')
```

```python
# get the class containing what we want
name2 = soup2.find(class_ = 'post-66316')
# get authors for second url
links2 = name2.find_all('a')
author2 = [item.text for item in links2][:41]
author2 = list(filter(lambda a: a != 'here' and a != ' here', author2)) 
author_last = author2[-3:]
author_replace = [author_last[0] + ", " + author_last[1], author_last[2]]
author2a = author2[:-3] + author_replace
author2a
```

```python
# Get the quotes of second url
quote_list2 = name2.find_all('li') # list of quotes
quote2 = [quote.text for quote in quote_list2][:-7]
quote2
```

#### Combine the author and quotes of the second page into a dataframe

```python
quotes_author2 = pd.DataFrame({
    'Quote': quote2,
    'Author': author2a
})
quotes_author2
```

## Combine the Results from both web pages

#### Bring the two dataframes together

```python
ds_quotes = pd.concat([quotes_author, quotes_author2], ignore_index=True)
ds_quotes
```

```python
ds_quotes.to_csv("data_science_quotes.csv", index=False)
```
