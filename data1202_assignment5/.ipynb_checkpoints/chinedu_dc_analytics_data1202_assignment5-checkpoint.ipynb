<b> Import Required Libraries </b>

```python
# Load required libraries
import glob
import numpy as np
import matplotlib.pyplot as plt
```

<b> Import all the datasets using glob </b>

```python
# Use glob to get all the datasets
dataset = sorted(glob.glob('data*.csv'))
dataset
```

<b> Create the `visualize` function </b>

```python
# Create a function to visualize the features in the given datasets

def visualize(dataset):
    """This function plots the average, maximum, and minimum of all features in the given data
    dataset = A csv file with all columns as numeric.
    """
    # read the dataset into a numpy array
    data = np.genfromtxt(dataset, delimiter = ",")
    x = np.arange(1,41)
    
    # set fig size
    plt.figure(figsize = (10, 20))
    fig, (ax1, ax2, ax3) =  plt.subplots(1,3)
    
    # set title for all plots in the subplot
    fig.suptitle(dataset[:-4])
    
    # plot the average value for all features in the dataset
    ax1.plot(x, data.mean(axis = 0))
    ax1.set(xlabel = "Feature #", ylabel = 'Average')
    
    # plot the maximum value for all features in the dataset
    ax2.plot(x, data.max(axis = 0))
    ax2.set(xlabel = "Feature #", ylabel = 'Max')
    
    # plot the minimum value for all features in the dataset
    ax3.plot(x, data.min(axis = 0))
    ax3.set(xlabel = "Feature #", ylabel = 'Min')
    
    # separate subplot and set subplot width
    fig.tight_layout()
    fig.set_figwidth(12)
    plt.show()
```

<b> Create the `identify_issues` function</b>

```python
# Create a function to identify issues with the datasets

def identify_issues(dataset):
    """This function identifies issues in the dataset
    dataset = A csv file with all columns as numeric.
    """
    dat = np.genfromtxt(dataset, delimiter = ",")
    dat_mean = dat.mean(axis = 0)
    dat_max = dat.max(axis = 0)
    dat_min = dat.min(axis = 0)
    if (dat_max[0] == 0 and dat_max[20] == 20):
        print('I doubt the integrity of this dataset\n\n')
    elif dat_min.sum() == 0:
        print('Why is this happening\n\n')
    else:
        print('Dataset looks good\n\n')
```

<b> Loop over all the datasets and apply the visualize and identify_issues function to each of the datasets </b>

```python
# for loop to loop over all elements of the dataset

for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
